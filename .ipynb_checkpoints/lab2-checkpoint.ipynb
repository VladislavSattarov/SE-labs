```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
#from ipywidgets import interact
```

```python
def omega(Gamma = 1., Beta = 1.):
    Alpha = np.linspace(0, 50, 1000)
    OMEGA_1 = np.sqrt(Alpha/2. + Gamma/2. + np.sqrt(((Alpha + Gamma)**2) /4. + Beta**2 - Alpha*Gamma))
    OMEGA_2 = np.sqrt(Alpha/2. + Gamma/2. - np.sqrt(((Alpha + Gamma)**2) /4. + Beta**2 - Alpha*Gamma))
    plt.plot(Alpha, OMEGA_1**2, color = 'red')
    plt.plot(Alpha, OMEGA_2**2, color = 'blue')
    plt.grid(True)
    plt.show()
#interact(interact_omega, B=(0, 20, 0.5), D=(0, 20, 0.5))
omega()
```

```python
def xt(Alpha = 1., Beta = 1., Gamma = 1., C1 = 1, C2 = 1):
    t = np.linspace(0, 20, 1200)
    OMEGA_1 = np.sqrt(Alpha/2. + Gamma/2. + np.sqrt(((Alpha + Gamma)**2) /4. + Beta**2 - Alpha*Gamma))
    OMEGA_2 = np.sqrt(Alpha/2. + Gamma/2. - np.sqrt(((Alpha + Gamma)**2) /4. + Beta**2 - Alpha*Gamma))
    x1 = C1 * np.cos(OMEGA_1 * t) + C2 * np.cos(OMEGA_2 *t)
    x2 = C1 * np.cos(OMEGA_1 * t) - C2 * np.cos(OMEGA_2 *t)
    plt.plot(t, x1, color = 'red')
    plt.plot(t, x2, color = 'blue')
    plt.grid(True)
    plt.show()
xt()
```

```python
 
t = np.linspace(-80, 80, 1000)
Alpha, Gamma = 3., 2.
Beta = 0.1
C1 = 10.5
C2 = 5.
OMEGA_1 = np.sqrt(Alpha/2. + Gamma/2. + np.sqrt(((Alpha + Gamma)**2) /4. + Beta**2 - Alpha*Gamma))
OMEGA_2 = np.sqrt(Alpha/2. + Gamma/2. - np.sqrt(((Alpha + Gamma)**2) /4. + Beta**2 - Alpha*Gamma))
x1 = C1 * np.cos(OMEGA_1 * t) + C2 * np.cos(OMEGA_2 *t)
x2 = C1 * np.cos(OMEGA_1 * t) - C2 * np.cos(OMEGA_2 *t)
plt.figure(figsize=(12, 5))
plt.plot(t, x1, color = 'red')
plt.plot(t, x2, color = 'blue')
plt.grid(True)
plt.show()
```
