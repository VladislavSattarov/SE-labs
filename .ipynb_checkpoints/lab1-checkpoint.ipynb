```python
def rhs(t, X):
    x, y = X
    return [y, x*x*x+x*x-2*x]
sol0 = solve_ivp(rhs, [0., 0.1], (0., 0.), method = 'RK45', rtol = 1e-12);
x0, y0 = sol0.y

sol1 = solve_ivp(rhs, [0., 0.1], (-2., 0.), method = 'RK45', rtol = 1e-12);
x1, y1 = sol1.y

sol2 = solve_ivp(rhs, [0., 0.1], (1., 0.), method = 'RK45', rtol = 1e-12);
x2, y2 = sol2.y
sol3 = solve_ivp(rhs, [0., 5], (-1., 0.), method = 'RK45', rtol = 1e-12);
x3, y3 = sol3.y
sol4 = solve_ivp(rhs, [5., 0], (-1., 0.), method = 'RK45', rtol = 1e-12);
x4, y4 = sol4.y

sol5 = solve_ivp(rhs, [0, 3.1], (2.5,-4), method = 'RK45', rtol=1e-12)
x5, y5 = sol5.y

sol6 = solve_ivp(rhs, [-10, 10], (1.3,1.3), method = 'RK45', rtol=1e-12)
x6, y6 = sol6.y

sol7 = solve_ivp(rhs, [-10, 10], (1+0.0005 , 0.00086603), method = 'RK45', rtol=1e-12)
x7, y7 = sol7.y

sol8 = solve_ivp(rhs, [10, -10], (1+0.0005 , 0.00086603), method = 'RK45', rtol=1e-12)
x8, y8 = sol8.y

sol9 = solve_ivp(rhs, [-10, 10], (1-0.0005 , 0.00086603), method = 'RK45', rtol=1e-12)
x9, y9 = sol9.y

sol10 = solve_ivp(rhs, [0, -10], (1-0.0005 , 0.00086603), method = 'RK45', rtol=1e-12)
x10, y10 = sol10.y

sol11 = solve_ivp(rhs, [-10, 0], (-2+0.00037796,0.00092582), method = 'RK45', rtol=1e-12)
x11, y11 = sol11.y

sol12 = solve_ivp(rhs, [0, -10], (-2+0.00037796,0.00092582), method = 'RK45', rtol=1e-12)
x12, y12 = sol12.y

sol13 = solve_ivp(rhs, [-10, 0], (-2-0.00037796,0.00092582), method = 'RK45', rtol=1e-12)
x13, y13 = sol13.y

sol14 = solve_ivp(rhs, [0, -10], (-2-0.00037796,0.00092582), method = 'RK45', rtol=1e-12)
x14, y14 = sol14.y

sol15 = solve_ivp(rhs, [0, -10], (-2+0.00037796, -0.00092582), method = 'RK45', rtol=1e-12)
x15, y15 = sol15.y

sol16 = solve_ivp(rhs, [-10, 0], (-2+0.00037796, -0.00092582), method = 'RK45', rtol=1e-12)
x16, y16 = sol16.y

sol17 = solve_ivp(rhs, [0, 3.1], (-3.5,4), method = 'RK45', rtol=1e-12)
x17, y17 = sol17.y

sol18 = solve_ivp(rhs, [0, 6.1], (-0.4,0.4), method = 'RK45', rtol=1e-12)
x18, y18 = sol18.y

```

```python
def eq_quiver(rhs, limits, N=16):
    xlims, ylims = limits
    xs = np.linspace(xlims[0], xlims[1], N)
    ys = np.linspace(ylims[0], ylims[1], N)
    U = np.zeros((N, N))
    V = np.zeros((N, N))
    for i, y in enumerate(ys):
        for j, x in enumerate(xs):
            vfield = rhs(0.0, [x, y])
            u, v = vfield
            U[i][j] = u
            V[i][j] = v
    return xs, ys, U, V
```

```python
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid(True)
limits=[(-3,3), (-3,3)]
xs, ys, U, V = eq_quiver(rhs, limits)
plt.quiver(xs, ys, U, V, alpha=1)
plt.plot(x0, y0, 'r.')
plt.plot(x1, y1, 'rx')
plt.plot(x2, y2, 'rx')

plt.plot(x3, y3, 'r-')
plt.plot(x4, y4, 'r-')
plt.plot(x5, y5, 'r-')
plt.plot(x6, y6, 'r-')
plt.plot(x7, y7, 'y-')
plt.plot(x8, y8, 'y-')
plt.plot(x9, y9, 'y-')
plt.plot(x10, y10, 'y-')
plt.plot(x11, y11, 'y-')
plt.plot(x12, y12, 'y-')
plt.plot(x13, y13, 'y-')
plt.plot(x14, y14, 'y-')
plt.plot(x15, y15, 'y-')
plt.plot(x16, y16, 'y-')
plt.plot(x17, y17, 'r-')
plt.plot(x18, y18, 'g-')
```
