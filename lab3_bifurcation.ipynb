```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def diagram(lambda_, x):
  x_array = []
  lambda_array = []
  for i in lambda_:
    for j in x:
      tmp = []
      x0 = j
      for z in range(400):
        x0 = i*np.sin(x0)
        if z > 300:
          tmp.append(x0)
      x_array.append(tmp)
      lambda_array.append(i)
  plt.xlabel('Î»')
  plt.ylabel('x*')
  plt.plot(lambda_array, x_array, 'r.')
```

```python
lambda_ = np.linspace(0, 5, 80)
x = np.linspace(0, 10, 4)
diagram(lambda_, x)
```
